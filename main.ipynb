## Tópicos Teóricos

Coeficiente de agrupamento (clustering coefficient): mede o grau com que os nós de um grafo tendem a agrupar-se.

1. Coeficiente de Clustering apenas na Projeção dos Hospitais, devido o fato do grafo Município Hospital ser Direcionado e Bipartido.
2. Componentes conectados só servem parada Gráficos Direcionados, logo, apenas o grafo Município Hospital terá essa medida.

3. Componentes Fortemente Conectadas (SCCs):

    1. Número de SCCs: 225
    2. Tamanho Médio das SCCs: 1.0
    3. Isso significa que não há subgrafos no grafo onde todos os nós dentro de uma SCC podem ser alcançados uns dos outros bidirecionalmente. Cada SCC contém apenas um nó, indicando que não há ciclos direcionados completos. Isso acontece, pois, nesse grafo, os nós apenas saem de Municipio e vão para Hospitais.

4. Componentes Fracamente Conectadas (WCCs):

    1. Número de WCCs: 6
    2. Tamanho Médio das WCCs: 37.5
    3. As WCCs mostram que há grupos de nós que estão conectados de alguma forma quando a direção das arestas é ignorada. Você tem uma grande WCC e várias menores, indicando que há uma grande componente conectada e algumas isoladas.

```python
correspondencia_pre_durante = gf.encontrar_clusters_correspondentes(clustersHospitaisPre, clustersHospitaisDurante)
correspondencia_durante_pos = gf.encontrar_clusters_correspondentes(clustersHospitaisDurante, clustersHospitaisPos)

print("Correspondência de Clusters do Período Pré para Durante:")
for cluster_pre, cluster_durante in correspondencia_pre_durante.items():
    print(f"Cluster Pré: {cluster_pre} -> Cluster Durante: {cluster_durante}")

print("\nCorrespondência de Clusters do Período Durante para Pós:")
for cluster_durante, cluster_pos in correspondencia_durante_pos.items():
    print(f"Cluster Durante: {cluster_durante} -> Cluster Pós: {cluster_pos}")
```

```python
diferencas_pre_durante = gf.calcular_diferencas_clusters(grafo_hospitaisPre, clustersHospitaisPre, grafo_hospitaisDurante, clustersHospitaisDurante, correspondencia_pre_durante)
diferencas_durante_pos = gf.calcular_diferencas_clusters(grafo_hospitaisDurante, clustersHospitaisDurante, grafo_hospitaisPos, clustersHospitaisPos, correspondencia_durante_pos)

print("Diferenças entre Clusters do Período Pré para Durante:")
for diff in diferencas_pre_durante:
    print(diff)

print("\nDiferenças entre Clusters do Período Durante para Pós:")
for diff in diferencas_durante_pos:
    print(diff)
```

```python
prePesos = sum(cluster['mudanca_peso_total'] for cluster in diferencas_pre_durante)
durantePesos = sum(cluster['mudanca_peso_total'] for cluster in diferencas_pre_durante)
posPesos = sum(cluster['mudanca_peso_total'] for cluster in diferencas_durante_pos)
totalPesos = prePesos + durantePesos + posPesos

print("Pré: ",prePesos)
print("Durante: ",durantePesos)
print("Pós: ",posPesos)
print("Total: ",totalPesos)
```
