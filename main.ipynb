## Tópicos Teóricos

Coeficiente de agrupamento (clustering coefficient): mede o grau com que os nós de um grafo tendem a agrupar-se.

1. Coeficiente de Clustering apenas na Projeção dos Hospitais, devido o fato do grafo Município Hospital ser Direcionado e Bipartido.
2. Componentes conectados só servem parada Gráficos Direcionados, logo, apenas o grafo Município Hospital terá essa medida.

3. Componentes Fortemente Conectadas (SCCs):

    1. Número de SCCs: 225
    2. Tamanho Médio das SCCs: 1.0
    3. Isso significa que não há subgrafos no grafo onde todos os nós dentro de uma SCC podem ser alcançados uns dos outros bidirecionalmente. Cada SCC contém apenas um nó, indicando que não há ciclos direcionados completos. Isso acontece, pois, nesse grafo, os nós apenas saem de Municipio e vão para Hospitais.

4. Componentes Fracamente Conectadas (WCCs):

    1. Número de WCCs: 6
    2. Tamanho Médio das WCCs: 37.5
    3. As WCCs mostram que há grupos de nós que estão conectados de alguma forma quando a direção das arestas é ignorada. Você tem uma grande WCC e várias menores, indicando que há uma grande componente conectada e algumas isoladas.

```python
def atribuir_indices_arestas(grafo):
    """
    Atribui índices únicos a cada aresta no grafo.

    Args:
    - grafo (nx.Graph): Grafo cujas arestas serão indexadas.

    Returns:
    - list: Lista de arestas com índices (formato: [índice, n1, n2, peso]).
    """
    arestas_com_indices = []
    for idx, (n1, n2, data) in enumerate(grafo.edges(data=True)):
        arestas_com_indices.append([idx, n1, n2, data['weight']])
    return arestas_com_indices


def comparar_arestas_com_indices(arestas_periodo1, arestas_periodo2):
    """
    Compara as arestas entre dois períodos, incluindo a variação dos pesos.

    Args:
    - arestas_periodo1 (list): Lista de arestas do período 1 (formato: [índice, n1, n2, peso]).
    - arestas_periodo2 (list): Lista de arestas do período 2 (formato: [índice, n1, n2, peso]).

    Returns:
    - dict: Dicionário com as arestas mantidas, removidas, adicionadas e variação dos pesos das arestas mantidas.
    """
    arestas_periodo1_dict = {(n1, n2): (idx, peso) for idx, n1, n2, peso in arestas_periodo1}
    arestas_periodo2_dict = {(n1, n2): (idx, peso) for idx, n1, n2, peso in arestas_periodo2}

    arestas_periodo1_set = set(arestas_periodo1_dict.keys())
    arestas_periodo2_set = set(arestas_periodo2_dict.keys())

    arestas_mantidas = arestas_periodo1_set.intersection(arestas_periodo2_set)
    arestas_removidas = arestas_periodo1_set.difference(arestas_periodo2_set)
    arestas_adicionadas = arestas_periodo2_set.difference(arestas_periodo1_set)

    variacao_pesos = {
        (arestas_periodo1_dict[aresta][0], aresta): arestas_periodo2_dict[aresta][1] - arestas_periodo1_dict[aresta][1]
        for aresta in arestas_mantidas
    }

    return {
        'mantidas': [(arestas_periodo1_dict[aresta][0], *aresta) for aresta in arestas_mantidas],
        'removidas': [(arestas_periodo1_dict[aresta][0], *aresta) for aresta in arestas_removidas],
        'adicionadas': [(arestas_periodo2_dict[aresta][0], *aresta) for aresta in arestas_adicionadas],
        'variacao_pesos': variacao_pesos
    }
```

```python
# Atribuir índices às arestas
arestasPreIndice = atribuir_indices_arestas(grafo_hospitaisPre)
arestasDuranteIndice = atribuir_indices_arestas(grafo_hospitaisDurante)
arestasPosIndice = atribuir_indices_arestas(grafo_hospitaisPos)

# Comparar as arestas entre os períodos
comparacao_pre_durante = comparar_arestas_com_indices(arestasPreIndice, arestasDuranteIndice)
comparacao_durante_pos = comparar_arestas_com_indices(arestasDuranteIndice, arestasPosIndice)

# Exibir resultados

print("Quantidade de Arestas: ")
print("Pré: ",len(arestasPre))
print("Durante: ",len(arestasDurante))
print("Pós: ",len(arestasPos))

print("---------------------------------------------------------------------------------------------")
print("Comparação de Arestas (Pré - Durante):")
print("---------------------------------------------------------------------------------------------")

print("Arestas Mantidas:")
cont = 0
for idx, n1, n2 in comparacao_pre_durante['mantidas']:
    print(f"  ID: {idx}, Aresta: ({n1}, {n2})")
    cont += 1
print("Quantidade de arestas Mantidas:",cont)
print("---------------------------------------------------------------------------------------------")

print("\nArestas Removidas:")
cont = 0
for idx, n1, n2 in comparacao_pre_durante['removidas']:
    print(f"  ID: {idx}, Aresta: ({n1}, {n2})")
    cont += 1
print("Quantidade de arestas Removidas:",cont)
print("---------------------------------------------------------------------------------------------")

print("\nArestas Adicionadas:")
cont = 0
for idx, n1, n2 in comparacao_pre_durante['adicionadas']:
    print(f"  ID: {idx}, Aresta: ({n1}, {n2})")
    cont += 1
print("Quantidade de arestas Adicionadas:",cont)
print("---------------------------------------------------------------------------------------------")

print("\nVariação dos Pesos das Arestas Mantidas:")
for (idx, aresta), variacao in comparacao_pre_durante['variacao_pesos'].items():
    print(f"  ID: {idx}, Aresta: {aresta}, Variação de Peso: {variacao}")
print("---------------------------------------------------------------------------------------------")

print("\nComparação de Arestas (Durante - Pós):")
print("---------------------------------------------------------------------------------------------")

print("Arestas Mantidas:")
cont = 0
for idx, n1, n2 in comparacao_durante_pos['mantidas']:
    print(f"  ID: {idx}, Aresta: ({n1}, {n2})")
    cont += 1
print("Quantidade de arestas Mantidas:",cont)
print("---------------------------------------------------------------------------------------------")

print("\nArestas Removidas:")
cont = 0
for idx, n1, n2 in comparacao_durante_pos['removidas']:
    print(f"  ID: {idx}, Aresta: ({n1}, {n2})")
    cont += 1
print("Quantidade de arestas Removidas:",cont)
print("---------------------------------------------------------------------------------------------")

print("\nArestas Adicionadas:")
cont = 0
for idx, n1, n2 in comparacao_durante_pos['adicionadas']:
    print(f"  ID: {idx}, Aresta: ({n1}, {n2})")
    cont += 1
print("Quantidade de arestas Adicionadas:",cont)
print("---------------------------------------------------------------------------------------------")

print("\nVariação dos Pesos das Arestas Mantidas:")
cont = 0
for (idx, aresta), variacao in comparacao_durante_pos['variacao_pesos'].items():
    print(f"  ID: {idx}, Aresta: {aresta}, Variação de Peso: {variacao}")
```

# Curva de Modularidade

```python
from networkx.algorithms.community import modularity
import matplotlib.pyplot as plt

def calcular_melhor_resolucao(grafo, resolucoes):
    melhor_resolucao = None
    melhor_modularidade = -1
    historico_modularidade = []

    for resolucao in resolucoes:
        comunidades = list(nx.community.greedy_modularity_communities(grafo, resolution=resolucao))
        mod = modularity(grafo, comunidades)
        historico_modularidade.append((resolucao, mod))
        
        if mod > melhor_modularidade:
            melhor_modularidade = mod
            melhor_resolucao = resolucao

    return melhor_resolucao, melhor_modularidade, historico_modularidade

# Exemplo de uso
resolucoes = [0.5, 0.6, 0.7, 0.8, 0.9, 1, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9, 2]

melhor_resolucao, melhor_modularidade, historico_modularidade = calcular_melhor_resolucao(grafoTotal, resolucoes)
print(f"Melhor resolução: {melhor_resolucao}, Modularidade: {melhor_modularidade}")

# Plotar o histórico de modularidade para diferentes resoluções
resolucao_values, modularidade_values = zip(*historico_modularidade)
plt.plot(resolucao_values, modularidade_values, marker='o')
plt.xlabel('Resolução')
plt.ylabel('Modularidade')
plt.title('Modularidade para diferentes resoluções')
plt.grid(True)
plt.show()
```
